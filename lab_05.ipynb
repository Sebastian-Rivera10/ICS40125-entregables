```python
# Contar la cantidad de películas y series
content_type_counts = df['type'].value_counts()

# Crear el gráfico de barras
fig = plt.figure(figsize=(8, 6))
sns.barplot(x=content_type_counts.index, y=content_type_counts.values, palette='viridis')
plt.title('Distribución de Tipos de Contenido en Netflix')
plt.xlabel('Tipo de Contenido')
plt.ylabel('Cantidad de Títulos')
plt.show()
```

El gráfico muestra claramente que **las películas son el tipo de contenido predominante** en Netflix, superando en número a las series de televisión. Esto implica que los usuarios que prefieren películas encontrarán una mayor variedad y cantidad de opciones en la plataforma, mientras que aquellos que buscan series podrían tener un catálogo más limitado en comparación con las películas.

### 2. Histograma de años de lanzamiento

```python
# Crear un histograma de los años de lanzamiento
fig = plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=30, kde=True, palette='magma')
plt.title('Distribución de Títulos por Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Títulos')
plt.show()
```

El histograma de los años de lanzamiento muestra que la mayoría del contenido en Netflix fue lanzado en años más recientes, con un pico notable alrededor de 2017-2020. Hay un patrón claro de crecimiento en la cantidad de títulos lanzados a partir de los 2000, intensificándose en la última década. ### 3. Proporción de clasificaciones por edad (`rating`)

```python
# Contar la proporción de cada clasificación
rating_counts = df['rating'].value_counts(normalize=True)

# Crear un gráfico de barras para las clasificaciones
fig = plt.figure(figsize=(10, 7))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='viridis')
plt.title('Proporción de Clasificaciones por Edad en Netflix')
plt.xlabel('Clasificación')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.show()
```

El gráfico de barras de las clasificaciones por edad muestra que 'TV-MA' (contenido para adultos) y 'TV-14' (mayores de 14 años) son las clasificaciones más comunes, seguidas por 'TV-PG' y 'R'. ### 4. Número de títulos agregados por año

```python
# Convertir 'date_added' a formato de fecha, forzando errores a NaT
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Eliminar filas donde 'date_added' es NaT (fechas inválidas)
df.dropna(subset=['date_added'], inplace=True)

# Extraer el año de 'date_added'
df['year_added'] = df['date_added'].dt.year

# Contar el número de títulos agregados por año
titles_added_per_year = df['year_added'].value_counts().sort_index()

# Crear el gráfico de línea
fig = plt.figure(figsize=(12, 6))
sns.lineplot(x=titles_added_per_year.index, y=titles_added_per_year.values, marker='o')
plt.title('Número de Títulos Agregados a Netflix por Año')
plt.xlabel('Año')
plt.ylabel('Número de Títulos Agregados')
plt.grid(True)
plt.show()
```

El gráfico de la cantidad de títulos agregados por año muestra un crecimiento constante y acelerado del catálogo de Netflix, especialmente a partir del 2014-2015, alcanzando un pico alrededor de 2019-2020. Luego, se observa una disminución en 2021. Este patrón de crecimiento coincide con la expansión internacional de la plataforma, donde Netflix invirtió fuertemente en contenido original y la adquisición de derechos para mercados globales, lo que resultó en un aumento significativo de su oferta.

### 5. Heatmap de lanzamientos por año y mes

```python
# Extraer el mes de 'date_added'
df['month_added'] = df['date_added'].dt.month_name()

# Reordenar los meses para el heatmap
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
               'August', 'September', 'October', 'November', 'December']
df['month_added'] = pd.Categorical(df['month_added'], categories=month_order, ordered=True)

# Crear una tabla pivote para el heatmap
release_month_year = df.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)

# Crear el heatmap
fig = plt.figure(figsize=(15, 8))
sns.heatmap(release_month_year, cmap='YlGnBu', fmt='d', linewidths=.5, cbar_kws={'label': 'Número de Títulos Agregados'})
plt.title('Heatmap de Títulos Agregados por Año y Mes en Netflix')
plt.xlabel('Mes')
plt.ylabel('Año')
plt.show()
```

El heatmap de lanzamientos por año y mes revela patrones interesantes de estacionalidad. Se observa que los meses de **octubre, noviembre y diciembre** suelen tener un mayor número de estrenos, lo que podría indicar una estrategia de Netflix para capitalizar la temporada de fiestas y vacaciones. También se aprecian picos en **enero y julio/agosto** en algunos años. Esto sugiere que Netflix utiliza el final y principio de año, así como los meses de verano, para lanzar contenido nuevo y atraer a la audiencia. Los meses de primavera y principios de verano (marzo a junio) suelen tener una actividad ligeramente menor en comparación.

### 6. Duración de películas por género

```python
# Filtrar solo películas
movies_df = df[df['type'] == 'Movie'].copy()

# Extraer la duración en minutos de la columna 'duration'
# Eliminar ' min' y convertir a numérico
movies_df['duration_minutes'] = movies_df['duration'].str.replace(' min', '').astype(float)

# Considerar solo el primer género listado para simplificar el análisis
movies_df['main_genre'] = movies_df['listed_in'].apply(lambda x: x.split(',')[0] if pd.notna(x) else 'Unknown')

# Filtrar géneros con al menos un número mínimo de películas para una mejor visualización
# Por ejemplo, géneros con al menos 50 películas
genre_counts = movies_df['main_genre'].value_counts()
main_genres_to_plot = genre_counts[genre_counts >= 50].index
movies_df_filtered = movies_df[movies_df['main_genre'].isin(main_genres_to_plot)]

# Crear un boxplot de duración por género principal
fig = plt.figure(figsize=(15, 8))
sns.boxplot(x='main_genre', y='duration_minutes', data=movies_df_filtered, palette='Set3')
plt.title('Distribución de Duración de Películas por Género Principal')
plt.xlabel('Género Principal')
plt.ylabel('Duración (minutos)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

El boxplot de la duración de películas por género principal muestra que existen diferencias significativas en la duración media y la variabilidad entre géneros. Por ejemplo, géneros como 'Documentaries' o 'Stand-Up Comedy' a menudo presentan duraciones más cortas, mientras que 'Dramas' o 'Action & Adventure' tienden a tener duraciones medias más largas. La variabilidad dentro de cada género también es notable, indicando que hay películas excepcionalmente largas o cortas dentro de cada categoría. Los outliers (puntos fuera de los bigotes) sugieren películas que se desvían considerablemente de la duración típica de su género, como documentales muy extensos o dramas muy concisos.

### 7. Top 10 países con más producciones

```python
# Para la columna 'country', algunos títulos tienen múltiples países. Dividiremos y contaremos.
# Asegurarse de que no haya NaN antes de dividir
countries_df = df['country'].dropna().apply(lambda x: x.split(', ')).explode()

# Contar las ocurrencias de cada país y obtener el top 10
top_10_countries = countries_df.value_counts().head(10)

# Crear el gráfico de barras
fig = plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values, palette='plasma')
plt.title('Top 10 Países con Más Producciones en Netflix')
plt.xlabel('País')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

El gráfico muestra que **Estados Unidos domina abrumadoramente** el catálogo de Netflix en términos de producción de contenido, seguido a una distancia considerable por India y el Reino Unido. Otros países como Canadá, Francia y España también aparecen en el top 10. Esta distribución es en gran medida esperada, dado el origen y la expansión de Netflix, así como la preeminencia de la industria cinematográfica y televisiva estadounidense e india. La presencia de países como el Reino Unido y Canadá también se alinea con la producción en inglés y la cooperación con estudios de EE. UU.

### 9. Relación entre duración y año de lanzamiento

```python
# Asegurarse de que movies_df está disponible y actualizado desde el punto 6
# Filtrar solo películas (si no se hizo ya o si el DataFrame original fue modificado)
movies_for_duration_analysis = df[df['type'] == 'Movie'].copy()

# Extraer la duración en minutos de la columna 'duration'
# Eliminar ' min' y convertir a numérico, manejando errores si los hubiera
movies_for_duration_analysis['duration_minutes'] = movies_for_duration_analysis['duration'].str.replace(' min', '').astype(float, errors='ignore')

# Eliminar filas con valores NaN en 'duration_minutes' que podrían surgir de la conversión
movies_for_duration_analysis.dropna(subset=['duration_minutes'], inplace=True)

# Crear el gráfico de dispersión
fig = plt.figure(figsize=(14, 7))
sns.scatterplot(x='release_year', y='duration_minutes', data=movies_for_duration_analysis, alpha=0.6, hue='release_year', palette='viridis', size='duration_minutes', sizes=(20, 400))
plt.title('Relación entre Duración de Películas y Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid(True)
plt.legend(title='Año de Lanzamiento', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

El gráfico de dispersión de la duración de películas por año de lanzamiento muestra que, en general, **no hay una tendencia clara y uniforme de cambio en la duración estándar de las películas a lo largo del tiempo**. Sin embargo, se observa una mayor densidad de películas con duraciones más variadas en los años más recientes, lo que podría reflejar un aumento en la producción de contenido diverso.

Los **outliers** son interesantes; aunque la mayoría de las películas se agrupan en un rango de duración 'estándar' (entre 70 y 120 minutos, aproximadamente), se pueden identificar películas excepcionalmente cortas o largas en todos los períodos, aunque las películas muy largas (más de 180-200 minutos) parecen ser más escasas en las últimas décadas. La concentración de puntos en los años más recientes también indica el gran volumen de contenido añadido a Netflix en esos periodos, lo que hace que los patrones sean más densos y la variabilidad más visible.

### 8. Películas vs Series según género

```python
# Para la columna 'listed_in', un título puede tener múltiples géneros.
# Vamos a expandir los géneros y luego contar por tipo y género.
genres_df = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

# Contar la cantidad de películas y series por género
genre_type_counts = genres_df.groupby(['listed_in', 'type']).size().unstack(fill_value=0)

# Filtrar los géneros principales para una mejor visualización (ej. top 15 géneros)
top_genres = genre_type_counts.sum(axis=1).nlargest(15).index
genre_type_counts_filtered = genre_type_counts.loc[top_genres]

# Crear el gráfico de barras apiladas
fig = plt.figure(figsize=(15, 8))
genre_type_counts_filtered.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], ax=plt.gca())
plt.title('Cantidad de Películas y Series por Género')
plt.xlabel('Género')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Contenido')
plt.tight_layout()
plt.show()
```

El gráfico de barras apiladas revela que géneros como 'Dramas', 'Comedies' y 'Action & Adventure' tienen una proporción significativa de películas. Por otro lado, géneros como 'International TV Shows' y 'Crime TV Shows' están, como es de esperar, dominados por series de televisión. También se observa que algunos géneros tienen una mezcla más equilibrada, pero en general, se puede apreciar una clara inclinación hacia películas en géneros más tradicionales y hacia series en géneros asociados a formatos televisivos específicos.

### Desafío Final: Combinaciones de Género y Clasificación por Edad

```python
# Para el desafío final, necesitamos expandir los géneros y clasificaciones por separado.
# Primero, limpiar la columna 'rating' ya que vimos que tiene algunos valores no estándar.
df_cleaned_ratings = df[df['rating'].isin(df['rating'].value_counts().nlargest(10).index)].copy()

# Expandir la columna 'listed_in' (géneros)
genres_and_ratings_df = df_cleaned_ratings.assign(listed_in=df_cleaned_ratings['listed_in'].str.split(', ')).explode('listed_in')

# Crear una tabla pivote para contar las combinaciones de género y rating
genre_rating_counts = genres_and_ratings_df.groupby(['listed_in', 'rating']).size().unstack(fill_value=0)

# Filtrar géneros con un número mínimo de apariciones para un heatmap más legible (ej. top 20)
top_genres_for_heatmap = genre_rating_counts.sum(axis=1).nlargest(20).index
genre_rating_counts_filtered = genre_rating_counts.loc[top_genres_for_heatmap]

# Crear el heatmap
fig = plt.figure(figsize=(16, 10))
sns.heatmap(genre_rating_counts_filtered, cmap='YlGnBu', fmt='d', linewidths=.5)
plt.title('Heatmap de Combinaciones Frecuentes de Género y Clasificación por Edad')
plt.xlabel('Clasificación por Edad')
plt.ylabel('Género')
plt.tight_layout()
plt.show()
```

El heatmap de combinaciones de género y clasificación por edad nos ofrece una visión detallada del público objetivo de los distintos tipos de contenido en Netflix.

Para el **público adulto**, géneros como 'Dramas', 'International TV Shows', 'Action & Adventure', 'Comedies' y 'Thrillers' tienen una fuerte presencia en clasificaciones como 'TV-MA' (para adultos maduros) y 'TV-14' (para mayores de 14 años). Las clasificaciones 'R' (restringido) y 'NC-17' también se asocian predominantemente con estos géneros de contenido más serio y adulto.

En contraste, los géneros dirigidos al **público familiar o infantil** son claramente 'Children & Family Movies' y 'Kids' TV', que se asocian mayoritariamente con clasificaciones como 'TV-Y' (todas las edades), 'TV-Y7' (mayores de 7 años) y 'TV-G' (audiencia general). Géneros como 'Documentaries' y 'Stand-Up Comedy' muestran una distribución más variada, abarcando desde clasificaciones más generales ('TV-PG') hasta las de adulto ('TV-MA'), dependiendo de la temática específica del título.
