```python
import pandas as pd

# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/netflix_titles.csv')
df.head()
```


### Parte 1: Limpieza y preparación

1. Revisar y describir el dataset:

   * ¿Cuántas filas y columnas tiene?
   * ¿Qué tipos de datos hay?
   * ¿Cuántos valores nulos hay por columna?

2. Transformar la columna `date_added` a tipo fecha.

3. Crear columnas auxiliares con `assign`:

   * Año (`year_added`)
   * Mes (`month_added`)



```python
print('--- Información del Dataset ---')
df.info()

print('\n--- Valores Nulos por Columna ---')
print(df.isnull().sum())

print('\n--- Transformando date_added a tipo fecha ---')
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

df_cleaned = df.assign(
    year_added=df['date_added'].dt.year,
    month_added=df['date_added'].dt.month
)

print('\n--- DataFrame con nuevas columnas (primeras 5 filas) ---')
display(df_cleaned.head())
```

## Parte 2: Técnicas avanzadas de pandas

4. Utilizar `.loc` para seleccionar películas (`type == 'Movie'`) que fueron agregadas después del año 2018.

5. Utilizar `str.contains()` y `str.extract()`:

   * Filtrar títulos que contienen la palabra 'love' (sin distinguir mayúsculas/minúsculas).
   * Extraer la duración en minutos para las películas desde la columna `duration`.

6. Aplicar `explode()` sobre la columna `listed_in` para obtener una fila por cada género.

7. Obtener un top 10 de géneros más frecuentes utilizando `value_counts()`.

8. Aplicar `where()` y `mask()` para marcar las películas de más de 120 minutos como contenido largo en una nueva columna.

9. Utilizar `.loc` para filtrar películas que cumplen con:

   * Más de 100 minutos de duración.
   * Rating igual a `'R'`.
   * País igual a `'United States'`.

10. Utilizar `.style` para formatear visualmente el top 10 de películas más largas.

```python
# 4. Utilizar .loc para seleccionar películas (type == 'Movie') que fueron agregadas después del año 2018.
movies_after_2018 = df_cleaned.loc[
    (df_cleaned['type'] == 'Movie') &
    (df_cleaned['year_added'] > 2018)
]
print('--- Películas agregadas después de 2018 (primeras 5 filas) ---')
display(movies_after_2018.head())
```

```python
# 5.1 Filtrar títulos que contienen 'love' (sin distinguir mayúsculas/minúsculas).
love_titles = df_cleaned[df_cleaned['title'].str.contains('love', case=False, na=False)]
print('--- Títulos que contienen "love" (primeras 5 filas) ---')
display(love_titles.head())

# 5.2 Extraer la duración en minutos para las películas.
# Asegurarse de que solo se procesan las películas (type == 'Movie') y la duración contiene 'min'
df_cleaned['duration_minutes'] = df_cleaned.loc[
    (df_cleaned['type'] == 'Movie') & (df_cleaned['duration'].str.contains('min', na=False)),
    'duration'
].str.extract(r'(\d+)').astype(float)

print('\n--- DataFrame con "duration_minutes" (primeras 5 filas donde la duración está en minutos) ---')
display(df_cleaned[df_cleaned['duration_minutes'].notna()].head())

# 6. Aplicar explode() sobre la columna listed_in para obtener una fila por cada género.
df_exploded = df_cleaned.assign(listed_in=df_cleaned['listed_in'].str.split(', ')).explode('listed_in')
print('\n--- DataFrame después de explode en "listed_in" (primeras 5 filas) ---')
display(df_exploded.head())
```

```python
# 7. Obtener un top 10 de géneros más frecuentes utilizando value_counts().
top_10_genres = df_exploded['listed_in'].value_counts().head(10)
print('\n--- Top 10 Géneros Más Frecuentes ---')
display(top_10_genres)
```

```python
# 8. Aplicar where() y mask() para marcar las películas de más de 120 minutos como contenido largo en una nueva columna.
# Crear una nueva columna 'is_long_content' en df_cleaned.
df_cleaned['is_long_content'] = False # Inicializar con False

# Usar .where() para marcar películas de más de 120 minutos como True
df_cleaned['is_long_content'] = df_cleaned['is_long_content'].where(
    df_cleaned['duration_minutes'] <= 120, # Condición: duración menor o igual a 120 minutos
    True # Si la condición es False (duración > 120), asigna True
)

# También podemos usar .mask() para el mismo propósito
# df_cleaned['is_long_content'] = False
# df_cleaned['is_long_content'] = df_cleaned['is_long_content'].mask(
#     df_cleaned['duration_minutes'] > 120, # Condición: duración mayor a 120 minutos
#     True # Si la condición es True (duración > 120), asigna True
# )

print('\n--- DataFrame con la nueva columna "is_long_content" (primeras 5 filas con valores de duración) ---')
display(df_cleaned[df_cleaned['duration_minutes'].notna()].head())
```

```python
# 9. Utilizar .loc para filtrar películas que cumplen con:
#    * Más de 100 minutos de duración.
#    * Rating igual a 'R'.
#    * País igual a 'United States'.

filtered_movies_loc = df_cleaned.loc[
    (df_cleaned['duration_minutes'] > 100) &
    (df_cleaned['rating'] == 'R') &
    (df_cleaned['country'] == 'United States')
]

print('\n--- Películas filtradas por duración, rating y país (primeras 5 filas) ---')
display(filtered_movies_loc.head())
```

```python
# 10. Utilizar .style para formatear visualmente el top 10 de películas más largas.

# Filtrar solo películas y ordenar por duración_minutos de forma descendente
top_10_longest_movies = df_cleaned[
    (df_cleaned['type'] == 'Movie') &
    (df_cleaned['duration_minutes'].notna())
].sort_values(by='duration_minutes', ascending=False).head(10)

print('\n--- Top 10 Películas Más Largas (Formato Visual) ---')
display(
    top_10_longest_movies[
        ['title', 'duration', 'duration_minutes', 'listed_in', 'rating']
    ]
    .style.hide(axis='index') # Ocultar el índice para una vista más limpia
    .bar(subset=['duration_minutes'], color='#5fba7d') # Barra de color para la duración
    .set_caption('Top 10 Películas Más Largas en Netflix') # Título de la tabla
    .set_properties(**{'text-align': 'left'})
    .set_table_styles([{'selector': 'th', 'props': [('text-align', 'left')]}]).format({'duration_minutes': "{:.0f} min"})
)
```

```python
# 11. ¿Cuáles son las combinaciones más frecuentes de género y rating en el dataset?
# (Sugerencia: utilizar `value_counts` con `subset=["listed_in", "rating"]` después de aplicar `explode()`).

# Asegurémonos de que no haya valores nulos en 'listed_in' o 'rating' antes de agrupar
# Aunque df_exploded ya debería tener 'listed_in' limpio, 'rating' puede tener nulos.
# Filtramos las filas donde 'rating' no es nulo para este análisis.
combinations_genre_rating = df_exploded[df_exploded['rating'].notna()].value_counts(subset=['listed_in', 'rating']).head(10)

print('\n--- Top 10 Combinaciones Más Frecuentes de Género y Rating ---')
display(combinations_genre_rating)
```

### Pregunta Bonus 12: ¿Existen películas con el mismo nombre (`title`) pero con distinto año de lanzamiento (`release_year`)?

```python
duplicate_titles_by_year = df_cleaned.groupby('title')['release_year'].nunique()
duplicate_titles_by_year = duplicate_titles_by_year[duplicate_titles_by_year > 1]

if not duplicate_titles_by_year.empty:
    print('--- Títulos con el mismo nombre pero diferente año de lanzamiento ---')
    for title in duplicate_titles_by_year.index:
        print(f"Título: '{title}'")
        display(df_cleaned[df_cleaned['title'] == title][['title', 'release_year', 'type']])
else:
    print('No se encontraron películas con el mismo nombre pero diferente año de lanzamiento.')
```
